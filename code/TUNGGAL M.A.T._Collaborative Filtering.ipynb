# Input nama user yang akan direkomendasikan

```python
USER_NAME = 'Hania'
```

# Import Library

```python
import pandas as pd
from math import sqrt
```

# Import Dataset

```python
df = pd.read_csv('../data/film2.csv')
```

```python
df.head(3)
```

# Menjadikan nama user sebagai index

```python
df.set_index('Nama Anda', inplace=True)
```

# Transpose Dataframe

```python
df = df.transpose()
```

```python
df.head(3)
```

# Korelasi antar penonton film

```python
df_corr = df.corr(method ='pearson') 
df_corr.head(3)
```

# Korelasi antara user dan penonton lainnya

```python
user_corr = df_corr[[USER_NAME]].loc[(df_corr[USER_NAME]>0) & (df_corr[USER_NAME]<1)]
user_corr
```

***
korelasi yang dipilih hanya dengan penonton lain yang korelasinya positif, artinya memang berkorelasi dengan user
***

# Melihat rating yang diberikan oleh penonton yang berkorelasi dengan user

```python
user_cor_rat = df
user_cor_rat = user_cor_rat.loc[user_cor_rat[USER_NAME] == 0]
user_cor_rat = user_cor_rat.transpose()
user_cor_rat[USER_NAME] = user_corr
user_cor_rat = user_cor_rat.dropna()
user_cor_rat = user_cor_rat.drop([USER_NAME], axis=1)
user_cor_rat
```

***
Rating yang ditampilkan hanya rating dari film yang belum ditonton oleh user
***

# Prediksi rating tiap film yang belum ditonton user

```python
sim_rat = (user_cor_rat.transpose()).dot(user_corr)
rating = []
l = len(sim_rat)
for i in range(l):
    a = sim_rat.iloc[i,0] / user_corr[USER_NAME].sum()
    rating.append(a)
    #print(rating)

rating
apa = {'rating': rating} 
result = pd.DataFrame(apa)
result
```

***
Rating ini didapatkan dengan cara mengalikan rating yang diberikan tiap penonton lain dengan korelasinya dengan user dibagi dengan nilai total korelasi penonton lain yang berkorelasi dengan user
<br>
<br>
\begin{equation*}
RatingMovie1 = \frac{(CorelationPenonton1 * RatingMovie1OlehPenonton1)+(CorelationPenonton2 * RatingMovie1OlehPenonton2)...}{total(CorelationSemuaPenontonYangBerkorelasiDenganUser)}
\end{equation*}
***

# Menggabungkan movie yang belum ditonton user dan prediksi ratingnya

```python
col_user = user_cor_rat.transpose()
col_user = col_user.reset_index()
col_user = col_user[['index']]
col_user.columns = ['movie']
col_user
```

```python
final_result = col_user.join(result, how='outer')
final_result = final_result.sort_values(by = 'rating', ascending = False)
final_result = final_result[final_result.rating != 0]
final_result
```

***
Movie yang direkomendasikan untuk user adalah **Avengers: End Game** kemudian **Spiderman: Far From Home** kemudian **Captain Marvel**
***

# Kesimpulan
**Collaborative filtering (CF)** user based digunakan untuk memberi rating pada movie yang belum ditonton oleh user, kemudian movie dengan rating tertinggi akan direkomendasikan kepada user untuk di tonton
Nilai Rating didapatkan dari total perkalian antara rating yang diberikan oleh user lain kepada film yang belum ditonton oleh user dikalikan dengan nilai korelasi dari user dengan user lain tersebut, kemudian dibagi dengan total korelasi semua user lain dengan user
<br> untuk case ini, movie yang direkomendasikan untuk user adalah **Avengers: End Game** kemudian **Spiderman: Far From Home** kemudian **Captain Marvel**. Namun movie yang paling direkomendasikan untuk user adalah **Avengers: End Game** karena memiliki nilai prediksi rating yang paling besar.
