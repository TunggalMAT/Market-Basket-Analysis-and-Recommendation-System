# Input user yang akan dicari rekomendasinya

```python
USER_NAME = 'Hania'
```

# Import Library

```python
import pandas as pd
from math import sqrt
```

# ------------------------------------------- DATASET CF --------------------------------------------

# Import Dataset Colaborative Filtering

```python
df_cf = pd.read_csv('../data/film2.csv')
```

# Menjadikan nama penonton sebagai index

```python
df_cf.set_index('Nama Anda', inplace=True)
```

# Menyimpan nama film yang ada pada dataset

```python
film = list(df_cf.columns.values)
```

# Transpose dataset

```python
df_cf = df_cf.transpose()
df_cf.head(3)
```

# Movie yang sudah ditonton user

```python
mov_watched = df_cf.loc[df_cf[USER_NAME] != 0]
mov_watched=mov_watched.reset_index()
mov_watched = mov_watched['index'].tolist()
mov_watched
```

# Movie yang belum ditonton user

```python
mov_unwatched = df_cf.loc[df_cf[USER_NAME] == 0]
mov_unwatched=mov_unwatched.reset_index()
mov_unwatched = mov_unwatched['index'].tolist()
mov_unwatched
```

# ------------------------------------------- DATASET CB --------------------------------------------

# Import Dataset Content Based Filtering

```python
df_cb = pd.read_csv('../data/apa3.csv')
df_cb
```

# ------------------------------------------- PROSES CB --------------------------------------------

# Menjadikan nama movie sebagai index

```python
df_cb.set_index('movie', inplace=True)
df_cb
```

# Mengencoder kolom yang bertipe object

```python
from sklearn.preprocessing import LabelEncoder
#df['movie'] = LabelEncoder().fit_transform(df['movie2'])
df_cb['negara'] = LabelEncoder().fit_transform(df_cb['negara'])
df_cb['genre'] = LabelEncoder().fit_transform(df_cb['genre'])
df_cb.head()
```

# MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
col2 = df_cb.columns
scaler = MinMaxScaler().fit_transform(df_cb)
df3 = pd.DataFrame(scaler, columns = col2)
df3
```

***
MinMax Scaler digunakan untuk mentransform features menjadi nilai yang biasanya nilainya range 0 sampai 1
<br>Hal ini dimaksudkan agar nilai yang ada pada semua feature tidak berbeda jauh gapnya
***

# Consine Similarity

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

df_c = cosine_similarity(df3)
```

```python
df_sim = pd.DataFrame(df_c,columns = film)
df_sim['movie'] = film
df_sim.set_index('movie', inplace=True)
df_sim
```

***
Consine similarity digunakan untuk mencari similarity setiap movie berdasarkan feature feature yang ada pada setiap movie
<br> semakin mendekati 1 nilai similaritynya berarti kedua movie tersebut sangat berkorelasi 
***

# Similarity antara movie yang sudah dan belum ditonton

```python
df_sim = df_sim.drop(mov_unwatched, axis=1)
df_sim = df_sim.loc[mov_unwatched]
```

```python
df_sim
```

# Rata-Rata similarity

```python
df_sim ['mean'] = df_sim.mean(axis=1)
```

***
Rata-Rata disini merupakan rata rata similarity dari movie yang belum ditonton dengan movie movie yang sudah ditonton sebelumnya oleh user
***

# Hasil CB Filtering

```python
cb_result = df_sim[['mean']]
cb_result = cb_result.reset_index().sort_values(by='mean', ascending=False)
cb_result = cb_result.head(3)
cb_result
```

***
Hasil Dari **Content Based Filtering** adalah user direkomendasikan untuk menonton film **The Lion King** kemudian Aladin dan 	Avengers: End Game
***

# ------------------------------------------- PROSES CF --------------------------------------------

# Korelasi antar penonton film

```python
df_corr = df_cf.corr(method ='pearson') 
df_corr.head()
```

# Korelasi antara user dan penonton lainnya

```python
user_corr = df_corr[[USER_NAME]].loc[(df_corr[USER_NAME]>0) & (df_corr[USER_NAME]<1)]
user_corr
```

***
korelasi yang dipilih hanya dengan penonton lain yang korelasinya positif, artinya memang berkorelasi dengan user
***

# Melihat rating yang diberikan oleh penonton yang berkorelasi dengan user

```python
user_cor_rat = df_cf
user_cor_rat = user_cor_rat.loc[user_cor_rat[USER_NAME] == 0]
user_cor_rat = user_cor_rat.transpose()
user_cor_rat[USER_NAME] = user_corr
user_cor_rat = user_cor_rat.dropna()
user_cor_rat = user_cor_rat.drop([USER_NAME], axis=1)
user_cor_rat
```

# Prediksi rating tiap film yang belum ditonton user

```python
sim_rat = (user_cor_rat.transpose()).dot(user_corr)
rating = []
l = len(sim_rat)
for i in range(l):
    a = sim_rat.iloc[i,0] / user_corr[USER_NAME].sum()
    rating.append(a)
    #print(rating)

rating
apa = {'rating': rating} 
result = pd.DataFrame(apa)
result
```

***
Rating ini didapatkan dengan cara mengalikan rating yang diberikan tiap penonton lain dengan korelasinya dengan user dibagi dengan nilai total korelasi penonton lain yang berkorelasi dengan user
<br>
<br>
\begin{equation*}
RatingMovie1 = \frac{(CorelationPenonton1 * RatingMovie1OlehPenonton1)+(CorelationPenonton2 * RatingMovie1OlehPenonton2)...}{total(CorelationSemuaPenontonYangBerkorelasiDenganUser)}
\end{equation*}
***

# Menggabungkan movie yang belum ditonton user dan prediksi ratingnya

```python
col_user = user_cor_rat.transpose()
col_user = col_user.reset_index()
col_user = col_user[['index']]
col_user.columns = ['movie']
```

```python
cf_result = col_user.join(result, how='outer')
cf_result = cf_result.sort_values(by = 'rating', ascending = False)
cf_result = cf_result.head(3)
cf_result
```

***
Movie yang direkomendasikan oleh colaborative filtering adalah **Avengers: End Game**, **Spiderman: Far From Home**, **Captain Marvel**
***

# Hasil Rekomendasi Akhir

```python
final_result = pd.merge(cb_result, cf_result, how='inner', on='movie')
final_result = final_result[final_result.rating != 0]
```

```python
final_result = final_result['movie'].tolist()
print ('Film yang direkomendasikan kepada', USER_NAME,'adalah', final_result)
```

Hasil dari Collaborative Filtering dan Content Based Filtering akan digabungkan dan di lihat movie mana yang sama sama direkomendasikan oleh kedua algoritma tersebut. movie yang sama tersebutlah yang direkomendasikan untuk user.
<br> Pada case ini, movie yang direkomendasikan kepada user menggunakan hybrid filtering adalah **Avengers: End Game**
